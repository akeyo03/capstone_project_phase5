## DATA UNDERSTANDING

### A. Dataset Overview:
The generated dataset contains information about various travel-related attributes:

**Previous_Location**: The city and country where the traveler is coming from.

**Locations (city/country)**: The city and country where the traveler is headed.

**Budget (USD)**: A range of travel budgets for the trip, ranging from low to high values (e.g., 500-800 USD, 1500-3000 USD).

**Weather_condition**: The weather expected at the destination (e.g., Sunny, Cold, Rainy).

**Month**: The time of the year the trip is taking place.

**Distance (km)**: The distance between the previous location and the destination in kilometers.

**Gender**: The gender of the traveler.

**Accommodation**: The type of accommodation chosen (e.g., GuestHouse, Hotel, Airbnb).

**Languages**: Languages spoken at the destination.

**Travel_Companions**: The people the traveler is going with (e.g., Family, Couple, Colleagues).

### B. Key Insights from the Data:
Outliers: There may be outliers in the distance column (e.g., extremely long travel distances). These should be identified and addressed to avoid skewing the analysis.

Missing Values: The data has no missing values.

Feature engineering: This may apply where the countries and cities may need to be seperated into different columns.

Languages: Since the Languages column contains lists of languages, some analysis may require converting these lists into categorical variables for easier comparison.

### C. Data Analysis:
To better understand the data distribution, we can conduct the following analyses:

**i) Univariate Analysis**

**Budget**: Visualizing the distribution of travel budgets can help identify trends in spending habits (e.g., the most common budget range for travelers).

**Distance**: Analyzing the distance traveled may reveal insights into the average distance people travel depending on the season or destination.

**Accommodation**: Observing the preferences in accommodation types (e.g., Hotel, Airbnb) based on budget or destination.

**ii) Bivariate Analysis**

**Budget vs. Weather_condition**: This could show whether travelers tend to spend more during certain weather conditions (e.g., Cold trips may have a higher budget due to additional clothing or special accommodations).

**Distance vs. Accommodation**: This helps understand if travelers covering long distances tend to prefer certain types of accommodation.

**Month vs. Budget**: Analyzing if there are seasonal spikes in travel expenses.

**Languages vs. Travel_Companions**: To explore if travelers with specific companions (e.g., family or couple) prefer destinations with certain language availability.

### i) Loading the Datasets

```python
#Import neccessary  libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
travel_df= pd.read_csv('data/Travel_data.csv')
travel_df.head()
```

```python
#Rename the columns
travel_df.rename(columns={"Previous_Location": "Current_location",
                   "Locations (city/country)":"Destination",
                    'Budget (USD)':'Budget(USD)',
                    'Distance (km)':'Distance(km)'}, inplace=True)
travel_df.head()
```

```python
travel_df.info()
```

```python
travel_df.shape
```

## ii) Exploratory Data Analysis (EDA)

### 1.Univariate Analysis

#### a) Distribution of Travel Budgets

```python
travel_df['Budget(USD)'].value_counts()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=travel_df, x='Budget(USD)', palette='viridis')

plt.title('Distribution of Travel Budgets')
plt.xlabel('Budget (USD)')
plt.ylabel('Number of Travelers')
plt.xticks(rotation=45)
plt.show()
```

### Interpretation

1500-3000 USD: This category has the highest count, with 2482 travelers, suggesting that most travelers tend to allocate a higher budget for their trips.
800-1500 USD: The second-largest category with 1840 travelers, showing a moderate budget preference.
500-800 USD: This is the least common budget range, with only 678 travelers, indicating that fewer people opt for a lower-budget trip.

The majority of travelers seem to prefer trips with higher budgets. This could be because of more luxurious accommodations, long-distance travel, or expensive destinations.
There is a significant drop in the number of travelers in the 500-800 USD category, suggesting that fewer travelers opt for budget-conscious trips.

#### b) Distribution of Travel Distance

```python
travel_df['Distance(km)'].describe()
```

```python
#Plot the graph
plt.figure(figsize=(10, 6))
sns.histplot(travel_df['Distance(km)'], kde=True, bins=30, color='blue')

plt.title('Distribution of Travel Distance')
plt.xlabel('Distance (km)')
plt.ylabel('Frequency')
plt.show()
```

#### Interpretation
The histogram shows a smooth distribution with peaks around the 8000-10,000 km mark, reflecting the high concentration of trips in that distance range. The KDE line overlaid on the histogram will smooth the distribution, highlighting the most common travel distances.
The right tail of the distribution (long distances) is heavier, showing that there are some very long trips, but they are less frequent.
On the left side, we have fewer shorter trips, which corresponds to the lower frequency of trips below 1000 km in the dataset.

Most travelers in this dataset are taking long-distance trips, with a notable concentration around 8000-10,000 km.
Shorter and extremely long trips are less common, as evidenced by fewer travelers in those bins.
The distribution reflects a preference for international travel, given the large distances involved.

### f) Distribution of travels by Month

```python
# Count of travels by month
monthly_travel_count = travel_df['Month'].value_counts().sort_index()
monthly_travel_count
```

```python
# Set the style
sns.set(style="whitegrid")
plt.figure(figsize=(10, 6))

# Line plot to show the trend
sns.lineplot(x=monthly_travel_count.index, y=monthly_travel_count.values, marker='o', color='r')

# Add labels and title
plt.title('Trend of Travels per Month', fontsize=16)
plt.xticks(rotation=90, fontsize=12)
plt.xlabel('Month', fontsize=14)
plt.ylabel('Number of Travels', fontsize=14)

# Display the plot
plt.tight_layout()
plt.show()

```

#### Interpretation

The line plot shows clear seasonal travel patterns, with July, April, and December being the peak travel months, while September, October, and March see the lowest travel activity. Summer and year-end holidays drive the highest numbers of trips, while September and October are a notably quiet period. These may be due to the influence of activities or weather conditions.

### 2) Bivariate Analysis
### a) Accommodation Preferences Based on Budget
This analysis explores whether there is a relationship between the type of accommodation and the budget.

```python
budget_accommodation_counts = travel_df.groupby('Budget(USD)')['Accommodation'].value_counts()

# Display the grouped output
print(budget_accommodation_counts)
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=travel_df, x='Budget(USD)', hue='Accommodation', palette='coolwarm')

plt.title('Accommodation Preferences Based on Budget')
plt.xlabel('Budget (USD)')
plt.ylabel('Number of Travelers')
plt.xticks(rotation=45)
plt.legend(title='Accommodation Type')
plt.show()
```

#### Interpretation
Higher Budgets (1500-3000 USD):Travelers with a higher budget show nearly equal preference for all three accommodation types. This suggests that at higher price points, people are open to a variety of accommodation options, perhaps because of the greater availability of quality choices.

Medium Budgets (800-1500 USD):Similarly, travelers in this range also show nearly even distribution across accommodation types. GuestHouses are slightly preferred, which might indicate a balance between cost and comfort in this budget range.

Lower Budgets (500-800 USD):For lower-budget travelers, Hotels are the most popular, followed by Airbnb and GuestHouse. This might indicate that even budget-conscious travelers are opting for more traditional or easily accessible accommodations.

### b) Travel Companions Preferences Based on Budget

```python
budget_companion_counts = travel_df.groupby('Budget(USD)')['Travel_Companions'].value_counts()

# Display the grouped output
print(budget_companion_counts)
```

```python
# Scatter plot 
plt.figure(figsize=(10, 6))
sns.countplot(data=travel_df, x='Budget(USD)', hue='Travel_Companions', palette='coolwarm')

plt.title('Travel Companions Preferences Based on Budget')
plt.xlabel('Budget (USD)')
plt.ylabel('Number of Travelers')
plt.xticks(rotation=45)
plt.legend(title='Travel Companions')
plt.show()
```

#### Interpretation

The plot visualizes the relationship between travel budgets and companion preferences, revealing that higher budgets, i.e 1500-3000 USD, are predominantly associated with couples, who represent the largest group of travelers in this range. Friends and family also travel frequently together at higher budgets, while solo travelers and those traveling with colleagues are less common. 

In the mid-budget category (800-1500 USD), couples remain the most significant group, but overall counts decrease compared to higher budgets, with solo travel still being relatively low. 

Lower budgets (500-800 USD) see a further decline in group travel, highlighting that financial constraints may limit social travel experiences. Overall, the data suggests that as budgets increase, the likelihood of traveling with partners or friends also rises. Throughout the distribution, couples seems to dominate having the highest number in each category of the budget.

### c) Gender vs Travel Companions
Analyzing the relationship between gender and travel companions can reveal preferences based on gender.

```python
gender_companion_counts = travel_df.groupby('Gender')['Travel_Companions'].value_counts()

# Display the grouped output
print(gender_companion_counts)
```

```python
# Count plot of Gender vs. Travel Companions
plt.figure(figsize=(10, 6))
sns.countplot(data=travel_df, x='Gender', hue='Travel_Companions', palette='coolwarm')
plt.title('Travel Companions by Gender', fontsize=16)
plt.xlabel('Gender', fontsize=14)
plt.ylabel('Count', fontsize=14)
plt.legend(title='Travel Companions')
plt.tight_layout()
plt.show()
```

#### Interpretation
The count plot of travel companions by gender reveals that both female and male travelers predominantly prefer to travel as couples, with 769 females and 795 males choosing this option, indicating a strong inclination towards partnership during travel. Following this, traveling with friends is also popular among both genders, with 671 females and 663 males opting for social travel. Family travel remains balanced, with 542 females and 557 males, while solo travel is less common, featuring 385 females and 357 males. Lastly, traveling with colleagues is the least favored option for both genders, highlighting a preference for personal over work-related travel. Overall, these insights suggest that shared experiences with companions are more valued than solo or professional travel.

### e) Relationship between Month and Distance Travelled
Analyzing the distance traveled in different months can help identify seasonal travel patterns.


```python
# Group by Month and calculate descriptive statistics
distance_summary = travel_df.groupby('Month')['Distance(km)'].describe()
# Display the summary statistics
print(distance_summary)
```

```python
# Box plot of Month vs. Distance
plt.figure(figsize=(10, 6))
sns.boxplot(data=travel_df, x='Month', y='Distance(km)', palette='coolwarm')
plt.title('Distance Travelled by Month', fontsize=16)
plt.xlabel('Month', fontsize=14)
plt.ylabel('Distance (km)', fontsize=14)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

#### Interpretation
Based on the above, the analysis reveals significant insights into travel patterns. April and July have the highest entry counts (629 and 669, respectively), indicating their popularity for travel. March shows the longest average distance traveled at 9188.43 km, while October has the shortest at 7637.87 km. The variability in distances is notable, especially in April and March, as indicated by their wide interquartile ranges and standard deviations, suggesting diverse travel preferences during these months. The box plot visualizes this variability, highlighting that October and November have more consistent, shorter trips, while April and March feature a broader range of distances, including outliers. .
